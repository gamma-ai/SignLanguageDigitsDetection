```python
from keras.preprocessing import image
img_paths = glob.glob("Examples/*.JPG")

def path_to_tensor(img_path):
    img = image.load_img(img_path, target_size=(64, 64))
    x = image.img_to_array(img)
    return np.expand_dims(x, axis=0)

def paths_to_tensor(img_paths):
    list_of_tensors = [path_to_tensor(img_path) for img_path in img_paths]
    return np.vstack(list_of_tensors)

img_input = paths_to_tensor(img_paths)


print('image shape: #Classes->{}, Height->{}, Width->{}, #Channels->{}'\
      .format(img_input.shape[0],img_input.shape[1],img_input.shape[2],img_input.shape[3]))
```

Now that the images have been converted to 4-D arrays, the images can be inputed into a tensorflow CNN model to plot an output of Image(5) at feature map (1 and 2). To achieve creating the first layer in the network, filters and strides need to be defined, as well as the tensorflow variable placeholder and The Convolutional layer to fit the train image into. This tensorflow implentation is used for visualization purposes to thoroughly explain what is happening in each specified parameter in the network. 

    Filters are a set of vertical and horizontal lines applied to the CNN architecture using the Tensorflow implementation.

    X1 is the input mini batch 4D tensor[10,64,64,3].

    The filters consist of a vertical white line and a horizontal white line that gets applied to the network.
    
    The strides are also a 1D array consisting of 4 elements: 1,stride_height,stride_width,1.
    
    The PADDING parameter has 2 different categories ("SAME" or "VALID"):
            -VALID: DOES NOT use zero padding. Depending on the stride it could miss rows and columns in the data.
            
            -SAME: DOES USE zero padding if necessary. output neurons = input neurons/ stride (rounded). The zeros are added evenly if possible to the input.
     
     The implementation of the pooling layers are used to shrink the input image to reduce the possibility of out-of-memory errors as well as using the specified pooling layer to find the Average of the k-size or Max of the Ksize. The size pooling layers will hold


Even though the filters have been defined, A neural network will be able to find the appropriate filters on its own. Experimentation with Cross Validation parameters is also an option to find the best parameters for the model but can cause be time consuming.

```python
# act = 'sigmoid'
# batch_size = 75
# w=4
# h=4
# reg = l2(.00025) #regularization to use within layers for overfitting
# epochs = 30

# num_classes = Y.shape[1]
# input_shape = X.shape[1:]
# print(num_classes)
# print(input_shape)

act = 'sigmoid'
batch_size = 75
w,h=8,8
reg = l2(0.00025) #regularization to use within layers for overfitting
epochs = 15

num_classes = Y.shape[1]
input_shape = X.shape[1:]
print(num_classes)
print(input_shape)
```

```python
model = Sequential()

#convolution 1st layer
model.add(Conv2D(64, kernel_size=(w,h),
                 activation=act,
                 kernel_regularizer=reg,
                 input_shape=(input_shape)))
model.add(BatchNormalization())
model.add(Dropout(.15))
model.add(MaxPooling2D())

#convolution 2nd layer
model.add(Conv2D(32, kernel_size=(w,h), kernel_regularizer=reg,activation=act))
model.add(BatchNormalization())
model.add(MaxPooling2D())

#convolution 3rd layer
model.add(Conv2D(64, kernel_size=(w,h),kernel_regularizer=reg, activation=act))
model.add(BatchNormalization())
model.add(MaxPooling2D())
model.add(Dropout(.15))

#Fully connected 1st layer
model.add(Flatten())
model.add(Dense(124,kernel_regularizer=reg))
model.add(BatchNormalization())
model.add(Activation(act))
model.add(Dense(num_classes))
model.add(Activation('softmax'))
# from keras.callbacks import LearningRateScheduler
# def step_decay(epoch):
#     init_rate = .9
#     drop = .05
#     epoch_drop = 10.0
#     lrate = init_rate*math.pow(drop, math.floor((1+epoch)/epoch_drop))
#     print('current learning rate is {} at epoch {}'.format(lrate,epoch+1))
#     return lrate
# lrate = LearningRateScheduler(step_decay)
# c2 = [lrate]

c1 = keras.callbacks.EarlyStopping(patience=10, verbose=1)
callback_list = [c1]
adam = keras.optimizers.Adam(lr=0.0025,decay=.0025,epsilon=.00025)    
model.compile(optimizer =adam, loss = 'categorical_crossentropy', metrics = ['accuracy','mse','mae'])

# callbacks = [keras.callbacks.EarlyStopping(patience=20,verbose=1),callback_list]
history=model.fit(X, Y,validation_split = 0.36,batch_size=batch_size,epochs=epochs,verbose=1,callbacks=callback_list)
print("Finished Fitting Model")
```

IMPLENTING CNN USING KERAS AS WELL AS LEARNING RATE SCHEDULER DECAY WITH AN INTIAL LEARNING RATE OF .0075 based on the previous model which gave a score of 87% AND DECREASING IT BY .0025 EVERY 3 EPOCHS. THIS WELL HELP THE MODEL TRAVEL THROUGH DIFFERENT LEARNING RATES OUTPUTING A GENEROUS ACCURACY SCORE EACH EPOCH WHILE REDUCING OVERFITTING WITH THE SIGMOID ACTIVATION FUNCTION AS WELL AS USING THE DROPOUT IMPLEMENTATION. THE USE OF THE SIGMOID IMPLEMENTATION WILL BE NOTICED IN THE VALIDATION ACCURACY VS THE TRAINING ACCURACY,(WHERE THE VALIDATION ACCURACY GENERALLY HAS A HIGH SCORE THAN THE TRAINING, THIS IS THE OVERFITTING CONTROL IN ACTION..)

```python
model = Sequential()

#convolution 1st layer
model.add(Conv2D(64, kernel_size=(w,h),
                 activation=act,
                 kernel_regularizer=reg,
                 input_shape=(input_shape)))
model.add(BatchNormalization())
model.add(Dropout(.15))
model.add(MaxPooling2D())

#convolution 2nd layer
model.add(Conv2D(32, kernel_size=(w,h), kernel_regularizer=reg,activation=act))
model.add(BatchNormalization())
model.add(MaxPooling2D())

#convolution 3rd layer
model.add(Conv2D(64, kernel_size=(w,h),kernel_regularizer=reg, activation=act))
model.add(BatchNormalization())
model.add(MaxPooling2D())
model.add(Dropout(.25))

#Fully connected 1st layer
model.add(Flatten())
model.add(Dense(124,kernel_regularizer=reg))
model.add(BatchNormalization())
model.add(Activation(act))
model.add(Dropout(.25))
model.add(Dense(num_classes))
model.add(Activation('softmax'))
from keras.callbacks import LearningRateScheduler

def step_decay(epoch):
    initial_lrate = 0.0075
    drop = 0.025
    epochs_drop = 3.0
    lrate = initial_lrate * math.pow(drop, math.floor((1+epoch)/epochs_drop))
    print('current learning rate is {} at epoch {}'.format(lrate,epoch+1))
    return lrate
lrate = LearningRateScheduler(step_decay)
class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = []
        self.lr = []
 
    def on_epoch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
        self.lr.append(step_decay(len(self.losses)))
loss_history = LossHistory()
lrate = LearningRateScheduler(step_decay)
c2 = [lrate, loss_history]

# adam = keras.optimizers.Adam(lr=0.0025,decay=.0025,epsilon=.00025)    
model.compile(optimizer ='adam', loss = 'categorical_crossentropy', metrics = ['accuracy','mse','mae'])

# callbacks = [keras.callbacks.EarlyStopping(patience=20,verbose=1),callback_list]
history=model.fit(X, Y,validation_split = 0.36,batch_size=batch_size,epochs=epochs,verbose=1,callbacks=c2)
print("Finished Fitting Model")
```
